```python
import requests
import pandas as pd

df = pd.DataFrame()
y = int(input())
for x in range(157,y+1):
    url='https://mgeps.philgeps.gov.ph/cms-home-pages/reg-merchants?page={x}&direction=SupPlatinumRegistrations.id+DESC%2CSupplierRegDetails.red_approval_date+DESC'.format(x=x)
    html = requests.get(url).content
    df_list1 = pd.read_html(html)
    df1 = df_list1[-1]
    df = df1.append(df).reset_index(drop=True)
    df.to_csv("103-200.csv")
```
